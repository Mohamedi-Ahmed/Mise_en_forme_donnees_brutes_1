```python
import pandas as pd

# Spécifie le chemin du fichier CSV
path_to_file = './data/darkness-dataset.csv'

# Ouvrir le fichier en mode lecture
with open(path_to_file, 'r', encoding='utf-8-sig') as file:
    for line_number, line in enumerate(file, start=1):
        
        # En-tête du DataFrame
        if line_number == 1:
            # Créer les colonnes à partir de la première ligne
            columns = line.strip().split(",")
            columns = [col.replace('linkedin_id;;;;', 'linkedin_id') for col in columns]
            df = pd.DataFrame(columns=columns)  # Créer un DataFrame vide avec les colonnes

        else:
            # Diviser la ligne en éléments
            line_split = line.strip().split(",")

            # Remplacer les éléments non valides par None si la longueur est > 1
            row_data = [elem if len(elem) == 1 else None for elem in line_split]

            # Ajuster la taille de row_data pour correspondre au nombre de colonnes
            if len(row_data) < len(columns):
                row_data.extend([None] * (len(columns) - len(row_data)))  # Ajouter des None s'il manque des éléments
            elif len(row_data) > len(columns):
                row_data = row_data[:len(columns)]  # Tronquer s'il y a trop d'éléments

            # Ajouter la ligne modifiée au DataFrame
            df.loc[len(df)] = row_data
        
        # Condition d'arrêt pour la démonstration (on arrête après la 2ème ligne)


# Afficher le DataFrame pour vérifier le contenu
df
```

## 2. Nettoyer les données

```python
path_to_cleaned_file = './data/cleaned_data.csv'

with open(path_to_file, 'r', encoding='utf-8', errors='replace') as infile, open(path_to_cleaned_file, 'w', newline='', encoding='utf-8') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    for row in reader:
        cleaned_row = [col.replace('""""', '"').replace('"""', '').replace('""', '"') for col in row]
        writer.writerow(cleaned_row)
```

## 3. Charger et séparer les colonnes

```python
cleaned_data = pd.read_csv(path_to_cleaned_file, sep=';')

# Vérifier les colonnes disponibles
print(cleaned_data.columns)
```

```python
# Séparer la première colonne basée sur la virgule
cleaned_data_split = cleaned_data['id,name,country_code,position,city,current_company,about,posts,experience,url,people_also_viewed,educations_details,education,recommendations_count,avatar,languages,certifications,recommendations,volunteer_experience,courses,followers,connections,current_company_company_id,current_company_name,publications,patents,projects,organizations,input_url,linkedin_id'].str.split(',', expand=True)

# Afficher un aperçu des données séparées
cleaned_data_split.head()
```

```python
# Nettoyer les guillemets supplémentaires et les éléments JSON
cleaned_data_final = cleaned_data_split.applymap(lambda x: x.replace('"', '').replace("{", '').replace("}", '') if isinstance(x, str) else x)

# Afficher un aperçu des données nettoyées
cleaned_data_final.head()
```
